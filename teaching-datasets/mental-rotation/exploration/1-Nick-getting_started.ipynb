## Read in Mental Rotation Data

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('../data/MentalRotation.csv')
df.head()
```

Any nulls?

```python
df.isna().any()
```

How many of each value (categorical vs discretized vs continuous data)

```python
df.nunique()
```

```python
dd = df.groupby(['Subject', 'Angle'])['Time'].mean()
dd.unstack().head(7)
```

```python
alt.Chart(dd.reset_index()).mark_point().encode(
    x='Angle:Q', 
    y='Time:Q',
)
```

```python
dd = df.groupby(['Angle', 'Matching'])['Time'].mean().reset_index()
alt.Chart(dd.reset_index()).mark_line().encode(
    x='Angle:Q', 
    y='Time:Q',
    color='Matching:N',
)
```

```python
dd = df.groupby(['Angle', 'Matching'])['Time'].mean().reset_index()
alt.Chart(dd.reset_index()).mark_line().encode(
    x='Angle:Q', 
    y='Time:Q',
    color='Matching:N',
)
```

```python
dd = df.groupby(['Subject', 'Angle', 'Matching'])['Time'].mean().reset_index()
lines = alt.Chart(dd.reset_index()).mark_line().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)
errorbars = alt.Chart(dd.reset_index()).mark_errorbar().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)

errorbars + lines
```

```python
dd = df.groupby(['Subject', 'Angle', 'Matching'])['Time'].mean().reset_index()
lines = alt.Chart(dd.reset_index()).mark_line().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)
errorbars = alt.Chart(dd.reset_index()).mark_errorbar().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)

errorbars | lines
```

```python
dd = df.groupby(['Subject', 'Angle', 'Matching'])['Time'].mean().reset_index()
lines = alt.Chart(dd.reset_index()).mark_line().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)
errorbars = alt.Chart(dd.reset_index()).mark_errorband().encode(
    x='Angle:Q', 
    y='mean(Time):Q',
    color='Matching:N',
)

errorbars + lines
```
