### Read Data

```python
df = pd.read_csv('../data/MentalRotation.csv')
df.head()
```

## Get Distribution of Response Times

```python
px.box(x=df.Angle, y=df.Time)
```

```python
px.violin(x=df.Angle, y=df.Time)
```

```python
px.violin(x=df.Angle, y=df.Time, color=df.Angle)
```

```python
px.histogram(x=df.Time, color=df.Matching)
```

```python
px.histogram(
    x=df.Time, 
    facet_col=df.Matching, 
    facet_row=df.Sex, 
    color=df.Correct, 
    histnorm="percent", 
    nbins=25,
    title='Distribution of Response Times',
    width=600,
    height=400,
)
```

```python
px.histogram(
    data_frame=df,
    x='Time', 
    facet_col='Matching', 
    facet_row='Sex', 
    color='Correct', 
    histnorm="percent", 
    nbins=25,
    title='Distribution of Response Times',
    width=600,
    height=400,
    animation_frame='Age',

)
```
